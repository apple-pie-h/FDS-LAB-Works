##Question 1: WAP to find if the given number is prime or not.

```python
a=int(input()) #taking input from the user
if a>1:
  for i in range(2,a):
    if a%i==0:
      print("Not Prime")
      break
  else:
    print("Prime")
else:
  print("Not Prime")
```

##Question 2: WAP to check if the given number is palindrome or not.

```python
n=int(input())
s=n
r=0
while n!=0:
  a=n%10
  r=r*10+a
  n=int(n/10)
if r==s:
  print("Palindrome")
else:
  print("Not Palindrome")
```

##Question 3: WAP to check if the given number is Armstrong or not.

```python
n=int(input())
s=n
r=0
c= len(str(abs(n))) #counting the number of digits in the given number
while n!=0:
  a=n%10
  r+=pow(a,c)
  n=int(n/10)
if r==s:
  print("Armstrong")
else:
  print("Not Armstrong")
```

##Question 4: WAP to check if the given strings are anagram or not.

```python
def is_anagram(s1, s2):
    s1 = s1.replace(" ", "").lower()
    s2 = s2.replace(" ", "").lower()
    return sorted(s1) == sorted(s2)

print(is_anagram("listen", "silent"))  # True
print(is_anagram("hello", "world"))    # False
```

##Question 7: WAP to find a Fibonacci upto a given number.

```python
n = int(input("Enter number of terms: "))
a, b = 0, 1

for _ in range(n):
    print(a, end=" ")
    a, b = b, a + b
```

##Question 8: WAP to find GCD of two numbers.

```python
def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return a

num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

print("GCD =", gcd(num1, num2))
```

##Question 9: WAP to demonstrate exception handling in python.

```python
try:
    a = int(input("Enter a number: "))
    b = int(input("Enter another number: "))
    result = a / b
    print("Result:", result)

except ZeroDivisionError:
    print("Error: Division by zero is not allowed.")

except ValueError:
    print("Error: Please enter valid integers.")

else:
    print("Division successful.")

finally:
    print("Program execution completed.")
```
